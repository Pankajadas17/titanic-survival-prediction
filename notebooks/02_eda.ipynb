```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/processed/cleaned_train.csv")
df.head()
```

```python
df.shape
```

```python
df["Survived"].value_counts()
```

More passengers did not survive than survived, indicating a slightly imbalanced dataset.

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Distribution")
plt.xticks([0, 1], ["Did Not Survive", "Survived"])
plt.show()
```

The dataset shows more non-survivors than survivors, highlighting the harsh survival conditions of the Titanic disaster.

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.legend(["Did Not Survive", "Survived"])
plt.show()
```

Female passengers had a significantly higher survival rate compared to males, supporting the “women first” evacuation policy.

```python
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.legend(["Did Not Survive", "Survived"])
plt.show()
```

First-class passengers had a much higher survival rate, while third-class passengers faced the highest mortality, indicating socio-economic bias in evacuation.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Most passengers were young adults, with fewer elderly passengers onboard.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Survived", y="Age", data=df)
plt.xticks([0,1], ["Did Not Survive", "Survived"])
plt.title("Age vs Survival")
plt.show()
```

Survivors tend to be slightly younger, suggesting children and younger adults had better survival chances.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare vs Survival")
plt.show()
```

Passengers who paid higher fares generally had better survival rates, correlating strongly with passenger class.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

Survival is moderately correlated with passenger class, fare, and family-related features, indicating their importance for modeling.
