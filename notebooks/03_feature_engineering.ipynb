```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/cleaned_train.csv")
df.head()
```

FamilySize = SibSp + Parch + 1

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df[["SibSp", "Parch", "FamilySize"]].head()
```

This feature captures social support, not raw counts.

```python
df["IsAlone"] = 0
df.loc[df["FamilySize"] == 1, "IsAlone"] = 1

df[["FamilySize", "IsAlone"]].head()
```

Binary features often help linear models.

```python
df["Title"] = df["Name"].str.extract(r" ([A-Za-z]+)\.", expand=False)
df["Title"].value_counts()
```

Titles reflect:
Age group (Master = child)
Gender
Social status

```python
rare_titles = ["Lady", "Countess", "Capt", "Col",
               "Don", "Dr", "Major", "Rev",
               "Sir", "Jonkheer", "Dona"]

df["Title"] = df["Title"].replace(rare_titles, "Rare")
df["Title"] = df["Title"].replace({"Mlle": "Miss", "Ms": "Miss", "Mme": "Mrs"})

df["Title"].value_counts()
```

This improves generalization

```python
df.drop(columns=["Name"], inplace=True)
```

 drop because We extracted all useful info from Name.

```python
df.head()
```

```python
df.drop(columns=["Ticket"], inplace=True)
```

```python
bool_cols = df.select_dtypes(include="bool").columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.to_csv("../data/processed/featured_train.csv", index=False)
print("✅ Dataset is now fully numeric and model-ready.")
```
